```python
def collate_fn(batch):
    """
    Retrieve batched by concatenating batch and zone_num directions
    """
    x, y = zip(*batch)
    x = torch.cat(x)
    y = torch.cat(y).unsqueeze(1)
    
    return (x, y)

dataloader = DataLoader(dataset, num_workers=0, batch_size=4, shuffle=True, collate_fn=collate_fn)

zones, labels = next(iter(dataloader))
print(labels.shape)
print(zones.shape)
show_zones(zones)
```
