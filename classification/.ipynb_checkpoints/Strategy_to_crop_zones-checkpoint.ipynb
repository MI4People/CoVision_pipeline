## Can we crop the zones of all test kits with the same location coordinates?

```python
import sys
import yaml
import json
import argparse
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
import torch

data_dir = 'data_classification'
```

```python
# List of 2-line kits only
kits_id = ['ACON_Ag', 'DeepBlue_Ag', 'Paramount-Jinwofu_Ag']

# Store membrane paths for each kit (split into positive, and negative cases)
membrane_pos_paths = {}
membrane_neg_paths = {}
for kit in kits_id:
    # Read kit labels
    kit_dir = os.path.join(data_dir, kit)
    labels_df = pd.read_excel(os.path.join(kit_dir, 'labels.xlsx'), usecols=['Sample ID', 'Zone 1', 'Zone 2'], index_col='Sample ID')
    # Store filenames after splitting by positive or negative cases
    membranes_pos_filename = iter(labels_df[labels_df.sum(1) == 2].index)
    membranes_neg_filename = iter(labels_df[labels_df.sum(1) != 2].index)
    # Update dictionaries
    membrane_pos_paths[kit] = [os.path.join(kit_dir, 'membranes', f'{name}.jpg') for name in membranes_pos_filename]
    membrane_neg_paths[kit] = [os.path.join(kit_dir, 'membranes', f'{name}.jpg') for name in membranes_neg_filename]
    # Log sizes
    print(f'# membranes {kit}: {len(membrane_pos_paths[kit])} (pos), {len(membrane_neg_paths[kit])} (neg)')
```

```python
def resize_image(image, new_height):
    """
    Resize image to a a new height, preserving the original width/height ratio.
    """
    ratio = image.shape[1] / image.shape[0]
    new_width = int(new_height*ratio)
    image = cv2.resize(image, (new_width, new_height))
    return image
```

### Small sample

Display small sample of positive membranes together with the universal cropping zones to corroborate that the zones are been cropped correctly.

```python
# Universal horizontal cropping locations (same for all kits)
crop_locs = [0.1, 0.50, 0.9]
n_kits = len(kits_id)
n_imgs = 20  # Number of images to display
height = 480  # New height for resizing, common to all membranes

fig, axs = plt.subplots(n_kits, n_imgs, figsize=(n_imgs, 10))

for i, kit in enumerate(kits_id):
    # Display kit name as y label
    axs[i, 0].set_ylabel(kit, size='large')
    for j in range(n_imgs):
        
        # Read and resize membrane image (positive only)
        membrane = resize_image(cv2.imread(membrane_pos_paths[kit][j]), height)
        
        # Draw green horizontal lines at each cropping location to devide zones
        x1, x2 = 0, membrane.shape[1]
        for loc in crop_locs:
            y = int(loc*height)
            cv2.line(membrane, (x1, y), (x2, y), (0, 255, 0), thickness=2)

        ax = axs[i, j]
        # Hide X and Y axis tick and label marks
        ax.xaxis.set_tick_params(labelbottom=False)
        ax.yaxis.set_tick_params(labelleft=False)
        ax.set_xticks([])
        ax.set_yticks([])
        
        ax.imshow(membrane[:,:,::-1])
        
        
fig.tight_layout()
plt.show()
```

### Full set check

Now we check that the zones are correctly cropped in ALL examples.

```python
def cropped_zones_checker(kit_id, membrane_paths, crop_locs):

    # Number of images to display, row and columns
    n_imgs = len(membrane_paths)
    n_cols = min(n_imgs, 20)
    n_rows = (n_imgs//n_cols + 1) if n_imgs % n_cols else n_imgs//n_cols

    height = 480  # New height for resizing, common to all membranes

    fig, axs = plt.subplots(n_rows, n_cols, figsize=(n_cols, 4*n_rows))

    for i in range(n_rows):
        for j in range(n_cols):
            # Catch single row case
            if n_rows > 1:
                ax = axs[i, j]
            else:
                ax = axs[j]
            if n_cols*i+j < n_imgs:
                # Read and resize membrane image
                membrane = resize_image(cv2.imread(membrane_paths[n_cols*i+j]), height)

                # Draw green horizontal lines at each cropping location to devide zones
                x1, x2 = 0, membrane.shape[1]
                for loc in crop_locs:
                    y = int(loc*height)
                    cv2.line(membrane, (x1, y), (x2, y), (0, 255, 0), thickness=2)

                ax.axis('off')
                ax.imshow(membrane[:,:,::-1])
            else:
                ax.axis('off')

    plt.suptitle(f'{kit_id} ({n_imgs})', size='xx-large')
    fig.tight_layout()
    plt.show()
```

```python
# List of 2-line kits only
kits_id = ['ACON_Ag', 'DeepBlue_Ag', 'Paramount-Jinwofu_Ag']
# Universal horizontal cropping locations (same for all kits) 
crop_locs = [0.1, 0.50, 0.9]
```

```python
# Choose specific kit and membrane paths (positive or negative)
kit = kits_id[2]
membrane_paths = membrane_neg_paths[kit]

# Run checker
cropped_zones_checker(kit, membrane_paths, crop_locs)
```
