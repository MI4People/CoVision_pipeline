```python
import yaml
import argparse

# Read arguments from dict and store them as Namespace
with open('config_classification.yaml', 'r') as f:
    config_dict = yaml.safe_load(f)
data_args = argparse.Namespace(**config_dict['data_args'])
training_args = argparse.Namespace(**config_dict['training_args'])

print(data_args)
print(training_args)
```

```python
from trainer_classification import TrainerClassification
from transformations_classification import PreprocessMembraneZones, AugmentedMembraneZones 
from utils_classification.visualizer import plot_metrics
```

## Check missclassified

```python
import os
import torch
from model_classification import ClassificationModel
from inference_classification import run_inference_dataset
from dataset_classification import MembraneZonesDataset
```

```python
load_path = os.path.join(training_args.saved_models_dir, '2024-03-03_18.28.26_best.state')
state = torch.load(load_path, map_location='cpu')
model = ClassificationModel()
model.load_state_dict(state['model_state'])
```

```python
transformation = PreprocessMembraneZones()
trainset = MembraneZonesDataset(data_args, kit_id=training_args.kit_id, data_mode='train', transform=transformation)
valset = MembraneZonesDataset(data_args, kit_id=training_args.kit_id, data_mode='val', transform=transformation)
testset = MembraneZonesDataset(data_args, kit_id=training_args.kit_id, data_mode='test', transform=transformation)

outputs_train_df = run_inference_dataset(trainset, model)
outputs_val_df = run_inference_dataset(valset, model)
outputs_test_df = run_inference_dataset(testset, model)

misclassified_train = outputs_train_df[outputs_train_df['targets']!=outputs_train_df['predictions']]
misclassified_val = outputs_val_df[outputs_val_df['targets']!=outputs_val_df['predictions']]
misclassified_test = outputs_test_df[outputs_test_df['targets']!=outputs_test_df['predictions']]
misclassified_train_ndx = misclassified_train.index.tolist()
misclassified_val_ndx = misclassified_val.index.tolist()
misclassified_test_ndx = misclassified_test.index.tolist()

n_wrong_train = len(misclassified_train)
n_wrong_val = len(misclassified_val)
n_wrong_test = len(misclassified_test)
print(f'Number of train examples missclassified: {n_wrong_train}')
print(f'Number of train examples missclassified: {n_wrong_val}')
print(f'Number of test examples missclassified: {n_wrong_test}')
```
